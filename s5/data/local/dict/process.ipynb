```python
import numpy as np
import re
import librosa as libr
import wave  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open('textr.txt','r')
```

```python
dat = f.read()
f.close()
```

```python
dat[0:200]
```

```python
dat_row = np.asarray(re.split("\n",dat))
```

```python
wav_path = []
kaldi_text = []
for x in dat_row:
    id = re.findall(r"\([^\)]*",x)[0][1:]
    text = re.findall(r"\>[^\<]*",x)[0][1:].strip()
    wav_path.append(r"s5/data/train/wav/c{}.wav".format(id))
    kaldi_text.append("c"+id+" "+text+"\n")
wav_path = np.asanyarray(wav_path)
```
